Source: https://www.faa.gov/air_traffic/publications/atpubs/fs_html/appendix_a.html

The International Civil Aviation Organization (ICAO) uses a variety of codes to classify flight types, including S, N, G, M, and X.

Flight type codes:

* S: Scheduled air service
* N: Non-scheduled air transport operation
* G: General aviation
* M: Military
* X: Everything else
